Build the three candidate architectures from the same model factory.
Pretrained weights are disabled in this notebook to avoid downloads.


```python
from pathlib import Path
import sys

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT / "src"))

from jaksuhealth_ai.models import build_model, count_trainable_parameters
```

```python
architectures = ["unet", "unetplusplus", "deeplabv3plus"]
models = {
    name: build_model(
        architecture=name,
        encoder_name="resnet18",
        encoder_weights=None,
        in_channels=1,
        num_classes=6,
    )
    for name in architectures
}
{name: count_trainable_parameters(model) for name, model in models.items()}
```
