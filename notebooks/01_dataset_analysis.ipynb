# JaksuHealth Dataset Analysis

This notebook inspects a prepared grouped split (eye-level for AMD-SD by default). Run

```python
from pathlib import Path
import sys
import pandas as pd

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT / "src"))

summary_path = ROOT / "data" / "AMD-SD_Split" / "split_summary.csv"
summary = pd.read_csv(summary_path)
summary
```

```python
lesion_columns = [column for column in summary.columns if column.endswith("_pixels")]
summary.set_index("split")[lesion_columns].T.plot(kind="bar", figsize=(12, 5))
```
